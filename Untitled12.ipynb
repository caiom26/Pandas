```python
import pandas as pd
import seaborn as sb
```

```python
Base_BTG = pd.read_excel('/content/Base BTG (1).xlsx')
Assinatura=pd.read_excel('/content/Assinatura CVM 178 (1).xlsx')
```

```python

Base_BTG.info()
```

```python

Base_BTG.groupby("Faixa Cliente").Conta.nunique()
```

```python
Base_BTG['Faixa']=Base_BTG['Faixa'].str.replace('Acima de 10MM','Acima de 10M')
Base_BTG['Faixa']=Base_BTG['Faixa'].str.replace('MM','M')
Base_BTG['Faixa']=Base_BTG['Faixa'].str.replace(' ','_')
Base_BTG.groupby("Faixa").Conta.nunique()
```

```python
Base_BTG["Termo de Marcação na Curva"].fillna('Ñ assinou',inplace=True)
```

```python
Base_BTG['Termo de Marcação na Curva']=Base_BTG['Termo de Marcação na Curva'].str.replace('Não assinou','Não')
Base_BTG.groupby("Termo de Marcação na Curva").Conta.nunique()
Base_BTG=Base_BTG.rename(columns={"Faixa Cliente":"Faixa"})
Base_BTG.info()
ClientesImportantes=Base_BTG.query("Assessor=='RODRIGO FARIA MACHADO' and Faixa=='Acima_de_10M'")
ClientesImportantes.head()
```

```python
Base_BTG.groupby("Assessor")["PL Total"].agg("sum").sort_values(ascending=False).reset_index()
```

```python
Base_BTG.groupby("Assessor").Conta.nunique().sort_values(ascending=False).reset_index()
```

```python
Base_BTG["mes_suitability"]=Base_BTG["Vencimento Suitability"].dt.month
Base_BTG.head(4)
Base_BTG.groupby("mes_suitability").Conta.nunique()
```
